# Graph 1 : Quarterly Trend of the Average Internet Access by Total Velocidad Bajada

```python
# Create a new column 'AñoTrimestre'
df['AñoTrimestre'] = df['Año'].astype(str) + 'Q' + df['Trimestre'].astype(str)

# Sort the DataFrame by 'AñoTrimestre' in ascending order
df_sorted = df.sort_values(by='AñoTrimestre', ascending=True)

# Calculate the average 'Total_velocidad_bajada' by 'AñoTrimestre'
df_avg = df_sorted.groupby('AñoTrimestre')['Total_velocidad_bajada'].mean().reset_index()

# Calculate the rate of growth of the average 'Total_velocidad_bajada'
df_avg['Rate_of_Growth'] = df_avg['Total_velocidad_bajada'].pct_change() * 100

# Plotting
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 10))

# Trend Graph
ax1.plot(df_avg['AñoTrimestre'], df_avg['Total_velocidad_bajada'], marker='o')
ax1.set_ylabel('Promedio Total_velocidad_bajada')
ax1.set_title('Tendencia Trimestral del Promedio Accesos por Total Velocidad Bajada')

# Rotate x-axis labels by 90 degrees
ax1.set_xticklabels(df_avg['AñoTrimestre'], rotation=90)

# Create a table for the Rate of Growth
table = ax2.table(cellText=df_avg[['Rate_of_Growth']].values.round(2),
                  colLabels=['Rate of Growth (%)'],
                  rowLabels=df_avg['AñoTrimestre'],
                  loc='center')

# Adjust table layout
table.auto_set_font_size(False)
table.set_fontsize(10)
table.scale(1.2, 1.2)
ax2.axis('off')  # Hide axes for the table

# Adjust subplots layout
fig.tight_layout()

# Show the plot
plt.show()
```

El promedio del Total de velocidad de bajada muestra una tendencia creciente con tasas de crecimiento positivas en casi todos sus trimestres.
Se debe destacar una tasa de crecimiento de 4.01% en el cuarto trimestre del 2021 debido a la pandemia y la demanda que se genero por servicios de internet como compras en mercados digitales, canales de streaming como netflix o youtube, etc.

# Graph 4: Quarterly trend of the average download speed (Ranges)

# Table 2.2: Participacion of each access internet technology type (over Total_acceso_Tipotecnologia) 
Total_acceso_BABA=Total_acceso_Tipotecnologia,
We'll find how much revenue produce each technology type.

```python
# Calculate the percentages
df1['Banda ancha fija (%)'] = (df1['Banda ancha fija'] / df1['Total_acceso_BABA']).round(decimals=4)
df1['Dial up (%)'] = (df1['Dial up'] / df1['Total_acceso_BABA']).round(decimals=4)
df1['ADSL (%)'] = (df1['ADSL'] / df1['Total_Acceso_Tipotecnologia'])
df1['Cablemodem (%)'] = (df1['Cablemodem'] / df1['Total_Acceso_Tipotecnologia'])
df1['Fibra óptica (%)'] = (df1['Fibra óptica'] / df1['Total_Acceso_Tipotecnologia'])
df1['Wireless (%)'] = (df1['Wireless'] / df1['Total_Acceso_Tipotecnologia'])
df1['Otros (%)'] = (df1['Otros'] / df1['Total_Acceso_Tipotecnologia'])

# Select the desired columns in the new dataframe
new_df = df1[['Año', 'Trimestre', 'Periodo', 'Banda ancha fija (%)', 'Dial up (%)',
             'ADSL (%)', 'Cablemodem (%)', 'Fibra óptica (%)', 'Wireless (%)', 'Otros (%)']]

# Display the new dataframe
new_df

# Create a new dataframe by multiplying 'Ingresos (miles de pesos)' with each column in 'new_df'
new_df_multiplied = new_df.copy()
new_df_multiplied.iloc[:, 3:] = new_df_multiplied.iloc[:, 3:].mul(df1['Ingresos (miles de pesos)'], axis=0)

# Display the new dataframe
new_df_multiplied.to_csv('IngresosPorTecnologia.csv')
new_df_multiplied.head(3)
```

# Graph 9: Revenue trend by technology

```python
# Create a subset of the filtered DataFrame with the desired columns
trend_df = new_df_multiplied[['Año', 'Trimestre', 'Periodo', 'ADSL (%)', 'Cablemodem (%)', 'Fibra óptica (%)', 'Wireless (%)']]

# Set 'AñoTrimestre' as a combination of 'Año' and 'Trimestre' columns
trend_df['AñoTrimestre'] = trend_df['Año'].astype(str) + 'Q' + trend_df['Trimestre'].astype(str)

# Sort the DataFrame by 'AñoTrimestre' in ascending order
trend_df = trend_df.sort_values(by='AñoTrimestre', ascending=True)

# Calculate the growth rates for each column
growth_rates = trend_df[['ADSL (%)', 'Cablemodem (%)', 'Fibra óptica (%)', 'Wireless (%)']].pct_change() * 100

# Display the growth rates table
growth_rates_table = pd.concat([trend_df['AñoTrimestre'], growth_rates], axis=1)
growth_rates_table.columns = ['AñoTrimestre', 'ADSL Growth Rate (%)', 'Cablemodem Growth Rate (%)', 'Fibra óptica Growth Rate (%)', 'Wireless Growth Rate (%)']
growth_rates_table = growth_rates_table.dropna()  # Remove rows with NaN values
growth_rates_table.reset_index(drop=True, inplace=True)  # Reset index

###print(growth_rates_table)

# Plot the trend for each column
plt.plot(trend_df['AñoTrimestre'], trend_df['ADSL (%)'], marker='o', label='ADSL')
plt.plot(trend_df['AñoTrimestre'], trend_df['Cablemodem (%)'], marker='o', label='Cablemodem')
plt.plot(trend_df['AñoTrimestre'], trend_df['Fibra óptica (%)'], marker='o', label='Fibra óptica')
plt.plot(trend_df['AñoTrimestre'], trend_df['Wireless (%)'], marker='o', label='Wireless')

# Set plot labels and title
plt.xlabel('AñoTrimestre')
plt.ylabel('Ingreso por tecnologia')
plt.title('Trend of Ingresos por Technologies (2014-2022)')

# Rotate x-axis labels by 90 degrees
plt.xticks(rotation=90)

# Add legend
plt.legend()

# Show the plot
plt.show()

growth_rates_table
```

# Graph 18: Internet Acces by range of download speed

```python
# Calculate the sum of Cantidad by Provincia
sum_cantidad_by_provincia = melted_df.groupby('Provincia')['Cantidad'].sum().sort_values(ascending=True)

# Plotting the lollipop graph
plt.figure(figsize=(12, 6))
plt.hlines(y=sum_cantidad_by_provincia.index, xmin=0, xmax=sum_cantidad_by_provincia.values, color='skyblue')
plt.plot(sum_cantidad_by_provincia.values, sum_cantidad_by_provincia.index, "o", color='blue')

# Add sum values with separation inside the lollipop markers
offset = max(sum_cantidad_by_provincia.values) * 0.009
for i in range(len(sum_cantidad_by_provincia)):
    plt.text(sum_cantidad_by_provincia.values[i] + offset, sum_cantidad_by_provincia.index[i],
             round(sum_cantidad_by_provincia.values[i], 2), ha='left', va='center')

plt.xlabel('Suma Cantidad')
plt.ylabel('Provincia')
plt.title('Suma Cantidad por Provincia')
plt.show()
```

# Graph 19: average internet access by range of download speed

```python
# Calculate the sum of Cantidad by Provincia
average_cantidad_by_provincia = melted_df.groupby('Provincia')['Cantidad'].mean().sort_values(ascending=True)

# Plotting the lollipop graph
plt.figure(figsize=(12, 6))
plt.hlines(y=average_cantidad_by_provincia.index, xmin=0, xmax=average_cantidad_by_provincia.values, color='skyblue')
plt.plot(average_cantidad_by_provincia.values, average_cantidad_by_provincia.index, "o", color='blue')

# Add sum values with separation inside the lollipop markers
offset = max(average_cantidad_by_provincia.values) * 0.009
for i in range(len(average_cantidad_by_provincia)):
    plt.text(average_cantidad_by_provincia.values[i] + offset, average_cantidad_by_provincia.index[i],
             round(average_cantidad_by_provincia.values[i], 2), ha='left', va='center')

plt.xlabel('Promedio Cantidad')
plt.ylabel('Provincia')
plt.title('Promedio de acceso a internet (por velocidad) por Provincia')
plt.show()
```
