```python
df1 = pd.read_csv('csv_api\Internet_Penetracion.csv')
#Número de accesos al servicio de Internet fijo por cada 100 hogares por provincia
df1.shape
```

```python
df1['id'] = df1['Año'].astype(str) + df1['Trimestre'].astype(str) + df1['Provincia'].astype(str)
df1.head()
```

```python
# List of columns to remove decimal points
columns_to_remove_decimal = ['Accesos por cada 100 hogares']

# Loop through the columns and remove decimal points
for col in columns_to_remove_decimal:
    df1[col] = pd.to_numeric(df1[col].str.replace(',', '.'))

# Convert 'Año' and 'Trimestre' columns to the appropriate data types
df1['Año'] = pd.to_numeric(df1['Año'], errors='coerce')
df1['Trimestre'] = pd.to_numeric(df1['Trimestre'], errors='coerce')

# Verify the modified DataFrame
print(df1.head())
```

```python
df4 = pd.read_csv('csv_api\Internet_Accesos-por-velocidad.csv')
#Número de accesos al servicio de Internet fijo por velocidad de bajada en cada provincia (trimestral)
```

```python
df4['id'] = df4['Año'].astype(str) + df4['Trimestre'].astype(str) + df4['Provincia'].astype(str)
```

```python
# List of columns to remove decimal points
columns_to_remove_decimal = ['HASTA 512 kbps',
       '+ 512 Kbps - 1 Mbps', '+ 1 Mbps - 6 Mbps', '+ 6 Mbps - 10 Mbps',
       '+ 10 Mbps - 20 Mbps', '+ 20 Mbps - 30 Mbps', '+ 30 Mbps', 'OTROS',
       'Total']

# Convert the columns to string type before applying the replacement
df4[columns_to_remove_decimal] = df4[columns_to_remove_decimal].astype(str)

# Loop through the columns and remove decimal points and nan
for col in columns_to_remove_decimal:
    df4[col] = df4[col].str.replace('.', '')
    df4[col] = df4[col].replace('nan', '')
    df4[col] = pd.to_numeric(df4[col].str.replace('.', ''))

# Convert 'Año' and 'Trimestre' columns to the appropriate data types
df4['Año'] = pd.to_numeric(df4['Año'], errors='coerce')
df4['Trimestre'] = pd.to_numeric(df4['Trimestre'], errors='coerce')

# Verify the modified DataFrame
print(df4.head())
```

```python
df4.isna().value_counts()
```

```python
df4 = df4.fillna(df4['OTROS'].mean())
```

```python
df4.isna().value_counts()
```

```python
df_accpenvel = pd.merge(df1,df4,on='id' )

print(df_accpenvel.shape)
df_accpenvel.columns
```

```python
df_accpenvel.drop(columns=['Año_y', 'Trimestre_y', 'Provincia_y'],inplace=True)
```

```python
# Order columns
df_accpenvel = df_accpenvel[['id','Año_x', 'Trimestre_x', 'Provincia_x','Accesos por cada 100 hogares',
        'HASTA 512 kbps', '+ 512 Kbps - 1 Mbps', '+ 1 Mbps - 6 Mbps', '+ 6 Mbps - 10 Mbps',
       '+ 10 Mbps - 20 Mbps', '+ 20 Mbps - 30 Mbps', '+ 30 Mbps', 'OTROS',
       'Total']]
```

```python
df_accpenvel.head(3)
```

```python
df_accpenvel['id'] = df_accpenvel['id'].str.replace(' ','')
```

```python
df = pd.read_csv('csv_api\Internet_BAF-por-provincia.csv')
#Número de accesos al servicio de Internet fijo por banda ancha y banda angosta en cada provincia (trimestral)
df['id'] = df['Año'].astype(str) + df['Trimestre'].astype(str) + df['Provincia'].astype(str)
df = df.drop(columns=['Año','Trimestre','Provincia'])
df['id'] = df['id'].str.replace(' ','')
df['Dial up']= df['Dial up'].str.replace('- ', '')
df['Banda ancha fija']= pd.to_numeric(df['Banda ancha fija'].str.replace('.',''))
df['Dial up']= pd.to_numeric(df['Dial up'].str.replace('.','').replace('- ', ''))
df['Total']= pd.to_numeric(df['Total'].str.replace('.',''))

df
```

```python
df_accpenvel = pd.merge(df_accpenvel, df, on='id')
#df_accpenvel.drop(columns=['Banda ancha fija_y','Dial up_y','Total_x'])
df_accpenvel.rename(columns={'Año_x':'Año', 'Trimestre_x':'Trimestre', 'Provincia_x':'Provincia', 'Total_x':'Total_velocidad_bajada','Total_y':'Total_Bandaancha_bandaangosta'}, inplace=True)
df_accpenvel
```

```python
#df_accpenvel.dropna(inplace=True)
df_accpenvel.drop_duplicates(keep='first', inplace=True)

#GUARDARRR
df_accpenvel.to_csv('tabla1.csv', index=False)
```
